```python
import numpy as np
# Copy your Particle class here

class Particle (object):
    def __init__(self, x, y, z, vx, vy, vz,m):
        self.x = x
        self.y = y
        self.z = z
        self.vx = vx
        self.vy = vy
        self.vz = vz
        self.m = m
        
    def Kinetic_E (self):
        return (0.5 *self.m*(self.vx**2 +self.vy**2 + self.vz**2))

    def distance(self, other):
        dx = self.x - other.x
        dy = self.y - other.y
        dz = self.z - other.z
        return np.sqrt(dx**2 + dy**2 + dz**2)
    
    def new_position(self, dt):
        self.x = self.x + self.vx*dt
        self.y = self.y + self.vy*dt
        self.z = self.z + self.vz*dt
        return(self.x, self.y, self.z)
```

```python
class ChargedParticle(Particle):
    
    def __init__(self, mass, x, y, z, vx, vy, vz, charge):
    
        super().__init__(mass, x, y, z, vx, vy, vz)
        self.charge = charge
    
    def Charged_Interaction(self, other):
        separation = self.distance(other)

        if separation < 0.25:
            print("interaction")

            if self.charge == -other.charge:
                self.vx =0
                self.vy =0
                self.vz = 0
                other.vx = 0
                other.vy =0
                other.vz = 0
                self.charge = 0
                other.charge = 0
                print("merge")

            else:
                factor_self = -1 * (self.charge + other.charge) * (self.m / other.m)
                factor_other = -1 * (self.charge + other.charge) * (other.m / self.m)

                self.vx *= factor_self
                self.vy *= factor_self
                self.vz *= factor_self

                other.vx *= factor_other
                other.vy *= factor_other
                other.vz *= factor_other

                print("repel")
```

```python
dt = 0.05

p_1 = ChargedParticle(-5, -5, -5, 1, 1, 1, 1, 0.5)
p_2 = ChargedParticle( 5,  5,  5, -1, -1, -1, 1, -0.5)

for step in range(300):
    p_1.new_position(dt)
    p_2.new_position(dt)

    p_1.Charged_Interaction(p_2)

    print("step", step)
    print("Position 1:", p_1.x, p_1.y, p_1.z, "Velocity:", p_1.vx, p_1.vy, p_1.vz)
    print("Position 2:", p_2.x,  p_2.y, p_2.z, "Velocity:", p_2.vx, p_2.vy, p_2.vz)
```
